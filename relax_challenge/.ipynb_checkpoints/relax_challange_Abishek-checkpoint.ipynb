## Summary of the findings:

The dataset contains total of 207917 instances of login with 8823 unique users. Using the given criterion of adopted user, approximately 11% users are identified as adopted. 


 
Using the dataset, a Random Forest Classifier is developed to predict the adopted users. The feature importance shows that age of the account is the best predictor to predict the adopted user. 




From the results, it can be concluded that the users who are signed up earlier are more likely to be adopted when compared to the new users. The model is able to predict adopted users with f1 scores of 0.94 and ROC of 0.63. The model performance could be better if we collect some other relevant information such as education, demographic and financial information of the users. 



## Coding Section

```python

### Importing Basic Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from datetime import date 
from datetime import datetime 
from datetime import timedelta 
from sklearn.preprocessing import LabelEncoder
```

```python
df1 = pd.read_csv("takehome_user_engagement.csv")
df1['time_stamp']=pd.to_datetime(df1['time_stamp'], format='%Y-%m-%d %H:%M:%S')
df1.head()
```

```python
print("There are {} users and {} total logins in the dataset".format(df1.user_id.nunique(), len(df1)))
```

```python
df1['date']=df1['time_stamp'].dt.date
df1.head()
```

```python
#Determination of a adopted user
adopted_user=[]
for user_id in df1['user_id'].unique():
    # first drop duplicate days and sort by day
    logins=df1[df1['user_id']==user_id]['date'].drop_duplicates().sort_values()
    logins=logins.diff(3).dropna()
    # calculate how many days has passed for every 3 logins
    if len([date for date in logins if date<=timedelta(days=7)])>1:
        adopted_user.append(user_id)
```

```python
print("total adopted users are {} out of {} unique users".format(len(adopted_user), df1.user_id.nunique() ))
```

```python
df2 = pd.read_csv("takehome_users.csv")
df2.head()
```

```python
df2.describe().T
```

```python
df2['adopted']=['yes' if i in adopted_user else 'No' for i in df2['object_id'].values]
df2.head()
```

```python
plt.figure(figsize=(10,6))
ax=sns.countplot(x = "adopted", data = df2)
plt.ylim([0,12000])
for p in ax.patches:
        ax.annotate('{}%'.format(round(100*p.get_height()/len(df2),2)), (p.get_x()+0.2, p.get_height()+500))
```

### Data seems like unbalanced so we need to pick a correct accuracy matric to deal with it. 

## lets creature a new feature "age" from creation time

```python
df2['creation_time']=pd.to_datetime(df2['creation_time'])
age=datetime.now()-df2['creation_time']
df2['age']=age/np.timedelta64(1, 's')
```

### Check for missing values

```python
df2.info()
```

"last_session_creation_time" and "invited_by_user_id" contains missing values. Missing in "invited_by_user_id" should corresponds to the account that are not invited. So, let's impute the missing value by None for invited by user and mean value in last session creation time. Also, "object_id" and "org_id" should be converted to an object.

```python
df2['last_session_creation_time']=df2['last_session_creation_time'].fillna(df2['last_session_creation_time'].mean())
df2['invited_by_user_id']=df2['invited_by_user_id'].fillna('None')
```

```python
df2['object_id']=df2['object_id'].astype('object')
df2['org_id']=df2['org_id'].astype('object')
```

```python
df2.info()
```

There are too many unique values for "invited_by_user", we can't convert each input into dummy variables. We can convert this data set in to binary values based on if the user is invited by another user or not.

```python
df2["invited_by_user_id"] = ["no" if i == "None" else "yes" for i in df2["invited_by_user_id"].values] 
```

## Feature Engineering

```python
df2.columns
```

```python
## use only useful features
df3=df2[['creation_source','opted_in_to_mailing_list','enabled_for_marketing_drip', 'invited_by_user_id', 'adopted', 'age']]
df3
```

### Now, let's incode the categorical features

```python
le = LabelEncoder()
df3.loc[:,'adopted']= le.fit_transform(df3.loc[:,'adopted'])
df3.loc[:,'invited_by_user_id']= le.fit_transform(df3.loc[:, 'invited_by_user_id'])
df3.head()
```

### let's see inter-features correlation

```python
plt.figure(figsize=(18,8))
_=sns.heatmap(df3.corr(),annot=True)
```

```python
cat=df3.select_dtypes('object').columns
cat
```

```python
dummies=pd.get_dummies(df3['creation_source'],drop_first=True)
df3=pd.concat([df3,dummies], axis=1)
df3=df3.drop('creation_source', axis=1)
df3.head()
```

```python
### Test train split and scaling
X=df3.drop(['adopted'], axis=1).values
y=df3['adopted'].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

RF=RandomForestClassifier(random_state=0)# 
RF.fit(X_train,y_train)

# Predicting the Test set results
y_pred = RF.predict(X_test)
y_pred_train = RF.predict(X_train)

# Making the Confusion Matrix
from sklearn.metrics import classification_report,confusion_matrix
print('Test Data Metrics:')
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))


print('\n')
print('Train Data Metrics:')
print(confusion_matrix(y_train, y_pred_train))
print(classification_report(y_train, y_pred_train))


from sklearn.metrics import roc_auc_score
y_pred_prob = RF.predict_proba(X_test)[:,1]
y_pred_prob_train = RF.predict_proba(X_train)[:,1]
    
print("roc_auc (training): {0:.3f}".format(roc_auc_score(y_train, y_pred_prob_train)))
print("roc_auc (test): {0:.3f}".format(roc_auc_score(y_test, y_pred_prob)))
```

### Hyperparameter tuning

```python
param = {
    'n_estimators':list(np.arange(150,200,5)),
    'max_depth': list(np.arange(10,100,5))}
```

```python
from sklearn.model_selection import RandomizedSearchCV

clf = RandomForestClassifier(random_state=0)
gs = RandomizedSearchCV(
    estimator=clf, param_distributions=param, 
    n_iter=100,
    scoring='roc_auc',
    cv=10,
    refit=True,
    random_state=0,
    verbose=True)

# Fit the object to our data
gs.fit(X_train, y_train)

print(gs.best_score_)
print(gs.best_params_)
```

## Model tuning

```python
RF=RandomForestClassifier(random_state=0, max_depth = 10, n_estimators = 160)# 
RF.fit(X_train,y_train)

# Predicting the Test set results
y_pred = RF.predict(X_test)
y_pred_train = RF.predict(X_train)

# Making the Confusion Matrix
from sklearn.metrics import classification_report,confusion_matrix
print('Test Data Metrics:')
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))


print('\n')
print('Train Data Metrics:')
print(confusion_matrix(y_train, y_pred_train))
print(classification_report(y_train, y_pred_train))


from sklearn.metrics import roc_auc_score
y_pred_prob = RF.predict_proba(X_test)[:,1]
y_pred_prob_train = RF.predict_proba(X_train)[:,1]
    
print("roc_auc (training): {0:.3f}".format(roc_auc_score(y_train, y_pred_prob_train)))
print("roc_auc (test): {0:.3f}".format(roc_auc_score(y_test, y_pred_prob)))
```
